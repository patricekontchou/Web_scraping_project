```python
# This script read a csv file collected using scrapy and scrap additional info using selenium.

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

#initialize google chrome driver.
driver = webdriver.Chrome()

# Read links file and stre in a list
f = open('./Projects/Web_scraping/gdoor\gdoor\spiders\links.csv', 'r+')
links = f.readlines()

result = []

#file where collected info will be stored.
sector_f = open('sector.csv','w+')

lenght = []
for link in links[1:]:
    #use slicing to remove extract quotation that was breaking links
    driver.get(link.strip()[1:-1])
    try:
        company_button = driver.find_element_by_xpath('//div[@data-tab-type="overview"]')
        company_button.click()
    except Exception as e:
        print('unprocessed link \n:',link)
        continue
    
    try:
        company = driver.find_element_by_xpath('//div[@class="css-16nw49e e11nt52q1"]').text
        result.append(' '.join(company.split()[:-2]))
    except Exception as e:
        print('company not found link \n:',link)
        continue
    try:
        vals = [el.text for el in driver.find_elements_by_xpath('//span[@class="value"]')]
    except Exception as e:
        print('description not found  link \n:',link)
        continue
        
    result.extend(vals)
    
    #Join list per link procesed and write in the file.
    line = '\t'.join(result[:7])
    sector_f.write(line)
    sector_f.write('\n')
    vals = []
   
    #Save number of links process per page 
    lenght.append(len(result))
    result = []

driver.close()

sector_f.close()
```
